```python
from matplotlib import axis


def outlier_count(col, data = X_train):
    
    # calculate your 25% quatile and 75% quatile
    q75, q25 = np.percentile(data[col], [75, 25])
    
    # calculate your inter quatile
    iqr = q75 - q25
    
    # min_val and max_val
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    
    # count number of outliers, which are the data that are less than min_val or more than max_val calculated above
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    
    # calculate the percentage of the outliers
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    
    if(outlier_count > 0):
        print("\n"+15*'-' + col + 15*'-'+"\n")
        print('Number of outliers: {}'.format(outlier_count))
        print('Percent of data that is outlier: {}%'.format(outlier_percent))
    
    
        
```

```python
import numpy as np

class LinearRegressionXavierGD:
    def __init__(self, learning_rate=0.01, epochs=1000, beta =0.9 ):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weight = None
        self.bias = 0
        self.beta = beta

    def initialize_weights(self, n):
        #Xavier Initialization
        bound = np.sqrt(1/n)
        self.weights = np.random.uniform(-bound, bound, n)
    
    def fit(self, X_1train, y_1train ):
        vtw =0
        vtb = 0
        m, n = X_1train.shape
        self.initialize_weights(n)  # Xavier Initialization
        for _ in range(self.epochs):
            y_pred = self.predict(X_1train)
            
            error = y_pred - y_1train
            grad = self.gradient(X_1train, error)
            vtw = self.beta * vtw + (1- self.beta)*grad*self.weight
            self.weight -= self.weight - self.learning_rate*vtw
            self.bias -= (1 / m) * np.sum(error)


    def gradient(self, X_1train, error):
        m = X_1train.shape[0]
        self.weight = (X_1train.T @ error) / m
        return self.weight 
    
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias

    def r2_score(self, y_1test, y_pred):
        res = np.sum((y_1test - y_pred) ** 2)  # Sum of Squared Residuals
        tot = np.sum((y_1test - np.mean(y_1test)) ** 2)  # Sum of Squared Total
        return 1 - (res / tot)

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
import mlflow
import mlflow.sklearn
```

```python
# Import our custom model
from matplotlib.pylab import beta


def evaluate_model(model, X_1train, y_1train, X_1test, y_1test, model_name ,beta):
    """ Train the model, make predictions, and compute evaluation metrics """
    model.fit(X_1train, y_1train,beta)
    y_pred = model.predict(X_1test)

    mse = mean_squared_error(y_1test, y_pred)
    r2 = r2_score(y_1test, y_pred)

    print(f"{model_name} - MSE: {mse:.4f}, R²: {r2:.4f}")
    return mse, r2

# Set up MLflow tracking
mlflow.set_experiment("Car Price Prediction")

def run_experiment(model, X_1train, y_1train, X_1test, y_1test, params):
    """ Train model with MLflow tracking """
    with mlflow.start_run():
        model.fit(X_1train, y_1train)
        y_pred = model.predict(X_1test)

        mse = mean_squared_error(y_1test, y_pred)
        r2 = r2_score(y_1test, y_pred)

        # Log parameters and metrics
        mlflow.log_params(params)
        mlflow.log_metric("MSE", mse)
        mlflow.log_metric("R2", r2)

        # Log the model
        mlflow.sklearn.log_model(model, "model")

        print(f"Logged model with params: {params} - MSE: {mse:.4f}, R²: {r2:.4f}")
        return mse, r2
```

```python
mlflow.set_experiment("Car Price Prediction")
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures

def run_experiment(model, X_train, y_train, X_test, y_test, params):
    """ Train model with MLflow tracking """
    with mlflow.start_run():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)

        # Log parameters and metrics
        mlflow.log_params(params)
        mlflow.log_metric("MSE", mse)
        mlflow.log_metric("R2", r2)

        # Log the model
        mlflow.sklearn.log_model(model, "model")

        print(f"Logged model with params: {params} - MSE: {mse:.4f}, R²: {r2:.4f}")
        return mse, r2

# ✅ K-Fold Cross Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Convert DataFrames to NumPy Arrays
X_1train_np = np.array(X_1train)
y_1train_np = np.array(y_1train)

# Define different models to compare
models = {
    "Polynomial Regression": make_pipeline(PolynomialFeatures(degree=1), LinearRegression()),
    "Lasso Regression": Lasso(alpha=0.1),
    "Ridge Regression": Ridge(alpha=1.0),
    "Linear Regression (Zero Init, No Momentum)": LinearRegressionXavierGD(learning_rate=0.0001, epochs=2000, beta=0),
    "Linear Regression (Xavier, No Momentum)": LinearRegressionXavierGD(learning_rate=0.0001, epochs=2000, beta=0),
    "Linear Regression (Xavier, With Momentum)": LinearRegressionXavierGD(learning_rate=0.0001, epochs=2000, beta=0.9),
}

# Store results
results = {}

for model_name, model in models.items():
    mse_list = []
    r2_list = []

    for train_idx, val_idx in kf.split(X_1train_np):
        X_train_fold, X_val_fold = X_1train_np[train_idx], X_1train_np[val_idx]
        y_train_fold, y_val_fold = y_1train_np[train_idx], y_1train_np[val_idx]

        model.fit(X_train_fold, y_train_fold)
        y_pred = model.predict(X_val_fold)

        mse = mean_squared_error(y_val_fold, y_pred)
        r2 = r2_score(y_val_fold, y_pred)

        mse_list.append(mse)
        r2_list.append(r2)

    results[model_name] = {"MSE": np.mean(mse_list), "R²": np.mean(r2_list)}

# Convert results to DataFrame
results_df = pd.DataFrame(results).T

# Print results
print(results_df)
```
